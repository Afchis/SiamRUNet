```python
import json
with open('/storage/ProtopopovI/_data_/ytb_vos/train.json') as data_file:
    data_json = json.load(data_file)
```

```python
import numpy as np
```

```python
path = "/storage/ProtopopovI/_data_/ytb_vos/train/JPEGImages/"
vid = data_json['videos']
anno = data_json['annotations']
```

```python
my_list = []
for i, data in enumerate(anno):
    if data['category_id'] == 1:
        my_list.append({ # len(my_list) --> 827
            'person_id' : anno[i]['id'],
            'video_id' : anno[i]['video_id'],
            })
```

```python
def Rle_to_numpy(RLE, width, height):
    NOT_RLE = []
    for i, data in enumerate(RLE):
        if i % 2 == 0:
            x = 0
        else:
            x = 1
        for j in range(data):
            NOT_RLE.append(x)
    np_array = np.asarray(NOT_RLE)
    np_array = np_array.reshape(height, width).tolist()
    return np_array
```

```python
new_json = []
for i, data in enumerate(my_list):
    #####
#     if i == 2: break
        #####
    for j, anno_data in enumerate(anno):
        if anno_data['id'] == data['person_id']:
            for k in range(len(anno_data['bboxes'])):
                
                for ii, vid_data in enumerate(vid):
                    if vid_data['id'] == data['video_id']:
                        width = vid_data['width']
                        height = vid_data['height']
                        file_name = vid_data['file_names'][k]
                    
                bbox = anno_data['bboxes'][k]
                try:
                    center = [bbox[0]+bbox[2]/2, bbox[1]+bbox[3]/2]
                    segmentation = anno_data['segmentations'][k]['counts']
                except TypeError:
#                     print(k)
                    center = None
                    segmentation = None
                #print(i, data, len(anno_data['bboxes']), file_name, width, height, center, bbox)
            
                for kk in range(len(anno_data['bboxes'])):
                    if anno_data['bboxes'][kk] is not None:
                        target_index = kk
                        for ii, vid_data in enumerate(vid):
                            if vid_data['id'] == data['video_id']:
                                target_name = vid_data['file_names'][target_index]
                        target_bbox = anno_data['bboxes'][target_index]
                        target_center = [target_bbox[0]+target_bbox[2]/2, target_bbox[1]+target_bbox[3]/2]
                        target_segmentation = anno_data['segmentations'][target_index]['counts']
                        target_info = {
                            'target_name' : target_name,
                            'target_bbox' : target_bbox,
                            'target_center' : target_center,
                            'target_segmentation' : target_segmentation,
                        }
                        break
            
                new_json.append({
                    'person_id' : data['person_id'],
                    'video_id' : data['video_id'],
                    'file_name' : file_name,
                    'width' : width,
                    'height' : height,
                    'center' : center,
                    'bbox' : bbox,
                    'segmentation' : segmentation,
                    'target_info' : target_info,
                })
```

```python
for item in new_json:
    print(item['person_id'], item['video_id'], item['bbox'], item['file_name'], item['target_info']['target_name'])
```

```python
js = json.dumps(new_json)

# Open new json file if not exist it will create
fp = open('test.json', 'a')

# write to json file
fp.write(js)

# close the connection
fp.close()
```
